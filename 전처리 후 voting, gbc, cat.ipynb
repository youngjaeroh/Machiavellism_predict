```python
from catboost import CatBoostClassifier, Pool  
from sklearn.ensemble import GradientBoostingClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import VotingClassifier
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('train_x.csv') #인덱스 col=0도 check!
df_test=pd.read_csv('test_c.csv')
sub=pd.read_csv('sample_submission.csv') 
```

```python
train_y = df['voted']
df.drop(['voted'], axis=1,inplace=True)
```

```python
gbc = GradientBoostingClassifier()
cat = CatBoostClassifier()
gbm = LGBMClassifier()
```

```python
voting = VotingClassifier(
            estimators=[('cat', cat), ('gbc', gbc), ('gbm', gbm)],
            voting='soft')
voting.fit(df, train_y)
v_pred=voting.predict(df_test)
```

```python
suv=pd.DataFrame(v_pred, columns = ['voted'])
```

```python
sub['voted'] = suv['voted']
```

```python
sub.to_csv("subvoting.csv", index=False)
```
