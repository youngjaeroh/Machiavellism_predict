```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.manifold import TSNE
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test_x.csv")
submit = pd.read_csv("sample_submission.csv")
```

```python
x_train = pd.get_dummies(train)
x_train.drop('voted', axis=1, inplace=True)
y_train = train['voted']
```

```python
test = pd.get_dummies(test)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(test)
```

```python
y_pred
```

```python
suv=pd.DataFrame(y_pred, columns = ['voted'])
```

```python
submit['voted'] = suv['voted']
```

```python
submit
```

```python
submit.to_csv('suv.csv', index=False)
```

```python
model1=DecisionTreeClassifier(random_state=35)
clf=model1.fit(x_train, y_train)
```

```python
tree_pred=clf.predict(test)
```

```python
suvT=pd.DataFrame(tree_pred, columns = ['voted'])
```

```python
submit['voted'] = suvT['voted']
```

```python
submit.to_csv('suvT.csv', index=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
R_pred=modelR.predict(test)
```

```python
from catboost import CatBoostClassifier, Pool  
from sklearn.ensemble import GradientBoostingClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import VotingClassifier
```

```python
gbc = GradientBoostingClassifier()
cat = CatBoostClassifier()
gbm = LGBMClassifier()
```

```python
voting = VotingClassifier(
            estimators=[('cat', cat), ('gbc', gbc), ('gbm', gbm)],
            voting='soft')
voting.fit(x_train, y_train)
prediction1=voting.predict(test)
```

```python
submit['voted'] = prediction1
```

```python
submit.to_csv('onehotvoting.csv', index=False)
```
